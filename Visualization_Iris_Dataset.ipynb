**Q1 Plot Histograms using iris dataset**

```python
# Please write codes with observations
"""  
In order to compare features of the flower categories, let us draw histogram based on a single feature from each of the flower category. 
Data distributed in below histogram shows that petal length of Versicolor and Virginica flowers are the most similar.
"""
setosa = iris.loc[iris.species =='setosa']
versicolor = iris.loc[iris.species == 'versicolor']
virginica = iris.loc[iris.species == 'virginica']

fig, ax = plt.subplots(nrows = 2, ncols = 3, figsize=(12,7))
ax[0,0].hist(setosa['petal_length'], bins= 5, edgecolor = 'k')
ax[0,0].set_title('Petal Length of setosa flower')
ax[0,1].hist(versicolor['petal_length'], bins= 5, edgecolor = 'k', color= 'red')
ax[0,1].set_title('Petal Length of versicolor flower')
ax[0,2].hist(virginica['petal_length'], bins= 5, edgecolor = 'k',  color= 'yellow')
ax[0,2].set_title('Petal Length of virginica flower')
ax[1,0].hist(setosa['petal_width'], bins= 5, edgecolor = 'k')
ax[1,0].set_title('Petal Width of setosa flower')
ax[1,1].hist(versicolor['petal_width'], bins= 5, edgecolor = 'k', color= 'red')
ax[1,1].set_title('Petal Width of versicolor flower')
ax[1,2].hist(virginica['petal_width'], bins= 5, edgecolor = 'k',  color= 'yellow')
ax[1,2].set_title('Petal Width of virginica flower')
fig.tight_layout();
```

**Q5 Plot Boxplots using the iris dataset**

```python
# Please write codes with observations
"""
Below Box plots give us clear visualization about length and width of petals and sepals of all three variety of flowers.  
"""
df = sns.load_dataset('iris')
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize=(14,8))
sns.set_theme(style="whitegrid")
sns.boxplot(x='species', y = 'petal_length', data = df, ax = ax[0,0])
sns.boxplot(x='species', y = 'petal_width', data = df, ax = ax[0,1])
sns.boxplot(x='species', y = 'sepal_length', data = df, ax = ax[1,0])
sns.boxplot(x='species', y = 'sepal_width', data = df, ax = ax[1,1])
fig.show()
fig.tight_layout();
```

**Q13 Plot double Bar plot for the following data also add legends**

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

# Please write your code with observations
"""
Below double bar plot shows scores acquired by men and women in different groups. 
The height of bars give a clear visualisation about the scores of men and women in different groups. 
Women have achieved more scores in group 1 and 3 whereas men have achieved more scores in group 2, 4 and 5.
"""

x = np.arange(len(labels))
width=0.4

fig, ax = plt.subplots(figsize=(10,6))
bar1 = ax.bar(x - width/2, men_means, width, label='Men')
bar2 = ax.bar(x + width/2, women_means, width, label='Women')

ax.set_title('Scores by Gender', size=16)
ax.set_ylabel('Scores', size=14)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(loc='upper right');
ax.grid(b=None)

def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
            xy=(bar.get_x() + bar.get_width() / 2, height),
            xytext=(0, 2),  # 3 points vertical offset
            textcoords="offset points",
            ha='center', va='bottom')

autolabel(bar1)
autolabel(bar2)
```

**Q15 Plot KDE plots using the iris dataset**

```python
# Please write your code with observations

"""
Kernel Density Estimate or KDE as the name suggests gives us visualization of probability distribution. 
Here in below KDE plot we have compared sepal width and sepal length of all three variety of flowers.
The density of colors shows the density of data for each flower category.
"""

sns.kdeplot(setosa['sepal_length'], 
            setosa['sepal_width'], color='r', 
            shade=True, 
            cmap="Reds", shade_lowest=False) 
  
sns.kdeplot(virginica['sepal_length'],  
            virginica['sepal_width'], color='b', 
            shade=True, 
            cmap="Blues", shade_lowest=False)

sns.kdeplot(versicolor['sepal_length'],  
            versicolor['sepal_width'], color='g', 
            shade=True, 
            cmap="Greens", shade_lowest=False);
```
